Interactive analysis of weather data from 5 major cities around the world.

**Data Source**: Open-Meteo API  
**Cities**: Berlin, London, New York, Tokyo, Sydney  
**Time Period**: Last 7 days (hourly data)

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

# Set plotting style
plt.style.use('default')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (12, 6)

print("üì¶ Libraries loaded successfully!")
```

```python
# Load weather data
weather_df = pd.read_csv('../data/clean_weather.csv')

# Convert timestamp to datetime
weather_df['timestamp'] = pd.to_datetime(weather_df['timestamp'])

print(f"üìä Dataset loaded: {len(weather_df):,} records")
print(f"üåç Cities: {', '.join(weather_df['city'].unique())}")
print(f"üìÖ Date range: {weather_df['timestamp'].min().date()} to {weather_df['timestamp'].max().date()}")

# Display first few rows
weather_df.head()
```

```python
# Temperature distribution by city
plt.figure(figsize=(12, 6))
weather_df.boxplot(column='temperature_c', by='city')
plt.title('üå°Ô∏è Temperature Distribution by City')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show();
```
